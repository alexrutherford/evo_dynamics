## The General Idea

We will consider a game that a group of agents play based on a public goods game. The action that each agent takes is determined by their strategy. Their action, and the actions of everyone else they play with will determine their payoff. Each agent’s strategy may change in 2 different ways; either by imitation (copying a strategy with higher payoff) or mutation (randomly changing to a different strategy). We would like to know, for a given game with well defined sets of rules and payoffs, how many agents will be cooperating, defecting or in any other state?

## A Simple Example

First we will look at a simple example based on the paper of Sigmund et al (see Supplementary Information).

In this toy example, there is a population of M agents and each one chooses either to cooperate in a PGG (strategy x), to defect (strategy y) or not to participate at all (strategy z). If you do not participate, you receive a small, fixed reward. The strategies of the agents may change in 2 ways; (1) by random mutation or (2) by imitating a different strategy which gives a better payoff (see [Martin Novak's book](http://www.amazon.com/Evolutionary-Dynamics-Exploring-Equations-Life/dp/0674023382) for more explanation of mutation and imitation).

We will also assume that imitation is very strong; if a strategy gives a better payoff then it will always be copied. Conversely if a strategy is worse then it will never be copied.

We want to find a way to analytically calculate how many agents we expect to cooperate, defect and not participate if we observe the population for a long time. We know that if all agents are cooperating, if one agent mutates to defection then all the cooperating agents will eventually copy the defector since free riding gives a higher payoff. Once all the agents are defecting, they will all get nothing since no-one contributes to the PGG. Therefore, if a non-participant is introduced then all the defectors will change to non-participants (since these agents always get paid a small amount). In turn, once all the agents are not participating a mutant cooperator can take over by neutral drift.

At this point we should specify what happens if there are not enough people to play the PGG. The PGG requires at least 2 participants, so if there is a lone agent among non-participants who wants to play the PGG then she will be forced to be a non-participant and receive the non-participant payoff. As soon as there is someone else to play the PGG game with, this agent's payoff will be calculated as normal.

So we see a cycle from cooperation to defection to non-participation and back to cooperation. So it isn’t easy to see what the long time average will be since the population will move between all 3 strategies. This is where the idea of the eigenvector of the transition matrix comes in.

For any such system with well defined states we can write down the transition matrix, this is a matrix of probabilities such that the (i,j) entry is the probability that the system will move from state i→ j. Our ‘system’ will be all M agents and the states will be when all the agents have the same strategy. We will call the ‘homgeneous’ states of all agents cooperating X, all defecting Y and all non-participating Z. Therefore, the transition matrix for this game will be

```python
display(Math(r'T =\begin{pmatrix}1-T_{X,Y}-T_{X,Z}&T_{X,Y}& T_{X,Z}\\T_{Y,X}&1-T_{Y,X}-T_{Y,Z}&T_{Y,Z}\\T_{Z,X}&T_{Z,Y}&1-T_{Z,X}-T_{Z,Y}\\\end{pmatrix}'))
```

1. The probability that a random mutation introduces an agent with strategy $y=\mu_{X,Y}$  
2. The probability that this single mutant can invade the population and lead all agents to switch to strategy Y ; this is known as the fixation probability $(\rho_{X,Y})$

The following is a key assumption in our model so read it carefully and take some time to fully understand it. In this formulation we assume that the mutation rate is low, that is many rounds of the PGG must be played before we see a random mutation happen. This means that once a mutation event happens, the mutant strategy either takes over (fixation) leading to transition to a new homogeneous state or the mutant strategy changes back via imitation returning the system to the same homogeneous state before the next mutation event occurs. Therefore, at any given time, at most two strategies are present.



```python
Image(filename='images/out.png',width=300)
```

**Gradual fixation of one strategy to take over another**

Addressing (1), mutations occur in the population at a rate $\mu$ (on average we must wait for $\mu$ rounds of the PGG before a mutation happens). The resultant strategy is chosen from the d − 1 other strategies at random; a cooperator can only mutate to a defector or non-participant, it cannot ‘mutate’ to a cooperator. This gives a mutation probability

$$\mu_{X,Y}=\frac{\mu}{(d-1)}$$

For a simple system of d = 3 states X ,Y and Z representing cooperators, defectors and non-participants respectively, we can construct T . The $\mu$ is constant for each entry so we will drop it from T.

```python
%%latex
\[\renewcommand\arraystretch{2}T = \begin{pmatrix}1-\frac{1}{2}\rho_{X,Y}-\frac{1}{2}\rho_{X,Z}&\frac{1}{2}\rho_{X,Y}&\frac{1}{2}\rho_{X,Z}\\\frac{1}{2}\rho_{Y,X}&1-\frac{1}{2}\rho_{Y,X}-\frac{1}{2}\rho_{Y,Z}&\frac{1}{2}\rho_{Y,Z}\\\frac{1}{2}\rho_{Z,X}&\frac{1}{2}\rho_{Z,Y}&1-\frac{1}{2}\rho_{Z,X}-\frac{1}{2}\rho_{Z,Y}\\\end{pmatrix}\]
```

Here each row sums to 1; in other words if the system is in state X, it must either transition to state Y or Z or it must stay in state X.

This might look a bit daunting, but it is very simple. When there is a mutation, the strategy to which the agent mutates is chosen from the other strategies at random. So when the system is in the state of X(all cooperators), the next mutation will give rise to a defector or a non-participant with equal probability. In general the factor of $\frac{1}{2}$ corresponds to  $\frac{1}{d−1}$

The individual entries of T can be populated by simple arguments under the limit $s \rightarrow \infty$ so that a strategy with a superior payoff will always be imitated and an inferior payoff will not. There are in fact only 3 possible values for the fixation probabilities $\rho_{i,j}$

$\boldsymbol{\rho_{i,j}=0}$: If $P_{j}<P_{i}$ for a single mutant with strategy $j$, then the mutation cannot invade and the fixation probability is 0.

$\boldsymbol{\rho_{i,j}=1}$: If $P_{j}>P_{i}$ for a single mutant with strategy $j$, then the mutation is beneficial and induces transition to a homegenous state $j$.

$\boldsymbol{\rho_{i,j}=\frac{1}{2}}$: This is peculiar to a single cooperator attempting to invade non-participants. The non-participants receive a fixed payoff of $\sigma$ but a single cooperator will also receive a payoff $\sigma$ since she has no partner with which to participate in a PGG. At the next imitation event involving the mutant cooperator, the cooperator will have the opportunity to imitate a non-participant. Since the payoffs are identical, the cooperator will revert to a non-participant with probability 1 , but is equally likely 2 to convert a non-participant to cooperation under a neutral drift. Once two or more cooperators are present, this strategy is dominant and they invade with probability 1.

Our intuitive understanding of PGGs tells us that in the absence of punishment, free-riding always pays ($\rho_{X,Y}=1$) and that unilateral cooperation in the face of defection does not ($\rho_{Y,X}=0$). When cooperation is underway, it pays to participate ($\rho_{X,Z}=0$) and due to the argument above, cooperators are slow to take over non-participants ($\rho_{Z,X}=\frac{1}{2}$). Finally, if no-one is playing the PGG then something is better than nothing ($\rho_{Y,Z}=1$ and $\rho_{Z,Y}=0$). Therefore $T$ reduces to

```python
%%latex 
\[\renewcommand\arraystretch{2}{T=\begin{pmatrix}\frac{1}{2}&\frac{1}{2}&0\\0&\frac{1}{2}&\frac{1}{2}\\\frac{1}{4}&0&\frac{3}{4}\\\end{pmatrix}}\]
```

Now the final step is to get the dominant eigenvector of this matrix, this will tell us the time- stationary distribution of the strategies. There are n eigenvectors for an \(n\times n\) matrix, and each one has a corresponding eigenvalue. The eigenvector with the largest corresponding eigenvalue is known as the dominant eigenvector. The eigenvectors and eigenvalues can be calculated analytically, in this case it involves solving a 3rd order polynomial. If you have access to a symbolic programming language such as Mathematica then you can use this.

However there is another method for calculating dominant eigenvectors numerically which comes in useful when the matrix is very large e.g. the adjacency matrix for a social network of 1000 people. This is known as the power iteration method. It is extremely simple and we will use Python to implement it now.

```python
transMat=np.zeros(shape=(3,3))

transMat[0][0]=0.5
transMat[0][1]=0.5
transMat[1][1]=0.5
transMat[1][2]=0.5
transMat[2][0]=0.25
transMat[2][2]=0.75

dist=[1,0,0]
# Choose an initial distribution; all cooperators

for i in xrange(100): dist=np.dot(dist,transMat)
# Multiply distribution and matrix together
# until the distribution is stable
print dist
```

This gives us the stationary distribution of \([0.25, 0.25, 0.5]\). This makes sense as cooperators flow to defectors, defectors flow to non-participants and non participants slowly turn back into cooperators in a continuous cycle; thus more time is spent as non-participants. In general we are interested in the level of cooperation that is maintained i.e. the proportion of cooperators in the stationary distribution. We can create new strategies or alter the rules of the game and perform the same calculation of the stationary distribution to measure how ‘good’ each set of strategies is at promoting cooperation and avoiding defection.

You might wonder why we used an initial distribution of all cooperators here. The truth is that the initial distribution is not important, after a small number of iterations through the loop there will be no memory of the initial distribution. This is sometimes referred to as the ‘burn-in’ period. You can check this for yourself;

**First, compare the end result from using different initial distributions. Secondly, plot how the distribution changes with each multiplication and how it eventually converges.**

We choose 100 steps for convergence here. Because this is a small matrix the eigenvector doesn’t take long to converge; if it were a larger matrix then many more steps would be needed. However, to be sure that you do not stop too early you could add some condition that the distribution now remains nearly the same after successive applications of the transition matrix.

## Introducing Punishers 

Now we introduce a new strategy; pool punishers. These agents in state v cooperate in the public goods game and also pay a fixed cost towards a pool which pays to punish defectors. Each defector can now free-ride in the PGG but will be charged a cost proportional to the number of pool punishers. The reason for introducing punishers is that the payoff for defectors will be lower so that it is no longer preferable to free-ride.

If everyone is non-participating except for one agent (whether she is a cooperators, defector or a punisher) then there is no-one to play the PGG with, therefore all agents will get the non-participant payoff.

The transition matrix for the population playing this 4 strategy game is given by:

```python
%%latex
\[ \begin{pmatrix} T_{XX}&T_{XY}&T_{XZ}&T_{XV}\\T_{YX}&T_{YY}&T_{YZ}&T_{YV}\\T_{ZX}&T_{ZY}&T_{ZZ}&T_{ZV}\\T_{VX}&T_{VY}&T_{VZ}&T_{VV} \end{pmatrix}\]
```

Where



```python
%%latex
\[T_{ij}\begin{cases}\frac{1}{3}\mu\rho_{ij}& \text{if } i\neq j\\1-\frac{1}{3}\mu\sum_{k\neq i}\rho_{ik} & \text{if } i=j\end{cases}\]
```

We can now populate this matrix using our intuition as follows. We fill in the rows from the top

1. Free riding still pays if there is no punishment as before ($\rho_{X,Y}=1$), if everyone is participating in the PGG then it pays to join in ($\rho_{X,Y}=0$) and it is cheaper not to punish ($\rho_{X,V}=0$)  
2. If everyone is defecting there is no incentive to contribute ($\rho_{Y,X}=0$ and $\rho_{Y,V}=0$) and something is better than nothing ($\rho_{Y,Z}=1$)  
3. A population of non-participants can neutral drift to cooperators ($\rho_{Z,X}=\frac{1}{2}$) and since pool-punishers don't have to punish if no-one is playing they can also neutral drift to pool-punishers ($\rho_{Z,V}=\frac{1}{2}$)  
4. Cooperators do not have to pay to punish, so can take over punishers ($\rho_{VX}=1$), but defectors will be heavily punished and we assume that non-participants only get a small reward so punishers are better off playing the PGG even if they have to pay into the punishment pool.  

```python
%%latex
\[ \begin{pmatrix}\frac{2}{3}&\frac{1}{3}&0&0\\0&\frac{2}{3}&\frac{1}{3}&0\\\frac{1}{6}&0&\frac{2}{3}&\frac{1}{6}\\\frac{1}{3}&0&0&\frac{2}{3}\end{pmatrix}\]
```

## Explicit Calculation

Up until now we have only used intuition and so called ‘hand-waving’ arguments to calculate the transition matrix. This is easy when the imitation strength is very high; in this case we know that a worse strategy will never be imitated and a better one will always be imitated and as a result ρ will always be strictly 0 or 1. However we would like to be able to calculate the stationary distribution when imitation strength is low, to do this we now need to write down the payoffs of each strategy explicitly.

Recall that the payoff in a public goods game depends on the number of cooperators (X), the number of defectors (Y) playing the PGG, the multiplication (r) and the contribution (c). A cooperator receives

```python
%%latex 
\[c\frac{(rX)}{(X+Y)}\]
```

We recall that in this setup, we continuously select a random sample of N agents out of M to play a PGG. If we introduce Z non-participants, then there is a chance that a cooperator or a defector might be paired with N-1 non-participants. If this happens then everyone will get the small non-participant payoff $\sigma$. The probability that this will happen can be calculated using binomial coefficients

```python
%%latex
\[\frac{\binom{Z}{N-1}}{\binom{M-1}{N-1}}\]
```

The numerator represents, assuming 1 member of the sample of N agents is the cooperator or defector in question, the probability that the remaining N-1 other agents are selected only from the Z non-participants. The denominator then normalises this by the number of different ways that the remaining N-1 other agents in the sample can be chosen from the remaining M-1 other agents in the population regardless of their strategy. If there are less than N-1 non-participants then it is impossible for the rest of the sample to be made up of non-participants. In this case $Z < N - 1$ and the binomial coefficient $\binom{Z}{N-1}$ evaluates to zero.

Finally we can write down the full payoff for the defectors, cooperators and non-participants based on the public goods game and the probability of being paired with non-participants and receiving only σ. Note that these are expected payoffs, we assume that each agent knows the number of each type of strategy in the population and can calculate the probabilities of being paired with each of them.

```python
%%latex
\[P_{\sigma} = \frac{\binom{Z}{N-1}}{\binom{M-1}{N-1}}\]
\[P_{second} = \frac{\binom{M-Y-2}{N-2}}{\binom{M-2}{N-2}}\]
\[Y \textrm{ payoff} = (P_{\sigma}.\sigma) + (1-P_{\sigma}).r.c.\frac{ M- Z- Y}{ M- Z} - B(N-1)\frac{V}{M-1}\]
\[X\textrm{ payoff} = (P_{\sigma} \sigma) + (1-P_{\sigma}).c.\left(r.\frac{ M- Z- Y}{ M- Z}-1\right)\]
\[Z \textrm{ payoff} = \sigma\]
\[V \textrm{ payoff} = (P_{\sigma} \sigma) + (1-P_{\sigma}).\left(c. \left[r.\frac{ M- Z- Y}{ M- Z}-1 \right] - G \right)\]
```

As a reminder we list and describe each strategy below

**Cooperators (X):** Participate and contribute c to the PGG

**Defectors (Y):** Participate but do not contribute to the PGG

**Loners (Z):** Neither participate nor contribute to PGG

**Pool Punishers (V):** Participate and contribute to the PGG (cooperate) and pay a fixed a prior cost G toward a punishment pool (central authority), which will punish defectors if defectors appear.

When considering imitation it is common to use a logistic function to determine the probability that one strategy copies another. When considering the probability for an agent playing strategy B to imitate an agent playing strategy A we have

```python
%%latex
\[\frac{1}{1+\mathrm{exp}[-s(P_{A}-P_{B})]}\]
```

Where s is the imitation strength. This has the general shape shown below

```python
%matplotlib inline
plt.plot([1. / (1. + math.exp(-1.*(x/10.))) for x in range(-40,40)],label='s=0.1')
plt.plot([1. / (1. + math.exp(-1.*(x/5.))) for x in range(-40,40)],label='s=0.5')
plt.plot([1. / (1. + math.exp(-1.*(x/1.))) for x in range(-40,40)],label='s=1.0')
plt.legend()
plt.show()
```

We see that generally the function gives higher probabilities when strategy A pays better than strategy B. However, when s is smaller, the chances of copying a worse strategy increase. For this reason, s is sometimes associated with the likliehood of agents making mistakes since it is irrational to adopt a strategy with a lower payoff. When s tends to infinity, the logistic function becomes a perfect step function.

What we want is a way to calculate the probability that, once a mutant strategy is introduced, that this mutant is imitated by another agent, and then another and then another and so on until all M agents have been converted. We can take the expression for this from [[1](http://www.nature.com/nature/journal/v466/n7308/full/nature09203.html)] using the theory of birth-death processes [[2](http://www.amazon.com/Evolutionary-Dynamics-Exploring-Equations-Life/dp/0674023382)]

```python
%%latex
\[\rho_{k,l}=\frac{1}{1+\sum_{q=1}^{M-1} \Pi_{N_{l}=1}^{q}\frac{\tau_{l\rightarrow k (N_{l})}}{\tau_{k\rightarrow l (N_{l})}}}\]
```

Where \(M\) is the size of the population and the number of agents with strategy $k$ or $l$ respectively is given by $N_{k}$ and $N_{l}$ with $M=N_{k}+N_{l}$. Here $\tau_{l\rightarrow k (N_{l})}$ represents the probability that one of the $N_{k}$ players will convert to strategy $l$ via imitation. If you don't understand this equation completely then don't worry, we will just use it as a black box.

This transition probability for a single agent can be written explicitly when obeying a logistic imitation probability.

```python
%%latex
\[\tau_{l\rightarrow k}(N_{l})=\frac{N_{l}}{M}\frac{M-N_{l}}{M}\frac{1}{1+\exp{[-s(P_{k}-P_{l})]}}\]
```

Where \(s\) is the imitation strength and $P_{k}$ and $P_{l}$ are the payoffs of strategies $k$ and $l$ which depend on the number of $k$ and $l$ players. Thankfully the fixation probability simplifies to

```python
%%latex
\[\rho_{k,l}=\frac{1}{1+\sum_{q=1}^{M-1} \exp[-s\sum_{N_{l}=1}^{q}(P_{k}-P_{l})]}\]
```

Now, instead of setting the entries of our transition matrix to 1,0 or $\frac{1}{2}$ we will calculate the entries numerically before finding the eigenvector and stationary distribution. Note that we can now change parameters such as the imitation strength to see how this affects the equilibrium distribution.

We can write a simple python script to do this. First of all we define the parameters of the public goods game and import the modules that we will need



```python
v='POOL'
x='COOPERATE'
y='DEFECT'
z='LONER'
# Labels for strategies
M=100
N=5
s=0.001
r=3.0
c=1
sigma=1.0
G=0.7
B=0.7
# Define game parameters
payoffDict={}
# This is a dictionary so that key 'v' returns # the payoff function for pool punishers etc
# We will define this later.
strategies=['v','x','y','z']
```

Next we need to define payoff functions for each strategy when the agent is playing a given number of agents playing the other strategies (this code simply implements the payoff equations above).

```python
#################
def p_v(V,X,Y,Z): 
#################
# Payoff for pool punisher
  assert V+X+Y+Z==M,'Pv: DONT SUM TO M!!!!'+' '.join([str(n) for n in [V,X,Y,Z]])
  returnVal=0.0

  returnVal+=r*((M-Z-Y)/float(M-Z))-1
  returnVal*=c
  returnVal-=G
  returnVal*=(1.-((scipy.special.binom(Z,N-1))/(scipy.special.binom(M-1,N-1)))) 
# PGG payoff-pool payment
  returnVal+=sigma*(scipy.special.binom(Z,N-1)/scipy.special.binom(M-1,N-1))
# Chance of non cooperation
  return returnVal 

#################
def p_x(V,X,Y,Z): 
#################
# Payoff for cooperator
  assert V+X+Y+Z==M,'Px: DONT SUM TO M!!!!'
  returnVal=0.0

  returnVal+=r*((M-Z-Y)/float(M-Z))-1 
  returnVal*=c*(1.-((scipy.special.binom(Z,N-1))/(scipy.special.binom(M-1,N-1)))) 
  # PGG payoff
  returnVal+=sigma*(scipy.special.binom(Z,N-1)/scipy.special.binom(M-1,N-1))
  # Chance of non cooperation
  return returnVal 
#################
def p_y(V,X,Y,Z): 
#################
# Payoff for defector
  assert V+X+Y+Z==M,'Py: DONT SUM TO M!!!!'
  returnVal=0.0

  returnVal+=c*r*((M-Z-Y)/float(M-Z))
  returnVal*=(1.-((scipy.special.binom(Z,N-1))/(scipy.special.binom(M-1,N-1)))) 
  # PGG payoff
  returnVal+=sigma*(scipy.special.binom(Z,N-1)/scipy.special.binom(M-1,N-1))
   # Chance of non cooperation
  returnVal-=(B*(N-1)*V)/float(M-1) 
   # Pool punishment term
  return returnVal 

#################
def p_z(V,X,Y,Z): 
#################
# Payoff for non-participant
  return sigma
```

Next we want a general function to calculate $\rho_{i,j}$ as defined above, this is the rate of transition between a homogeneous state of i to a homogeneous state of j. This function is probably the most complex, but understand that it is using a sum to calculate the probability that a mutant is imitated once, then again and so on for all M agents.

```python
#################
def getNumbers(arg0,arg1,n):
#################
    '''For a given pair of strategies (strings) returns full 
    tuple describing number of each strategy (int)'''
    vv=zz=xx=yy=0

    if arg0=='x':xx=M-n
    elif arg0=='y':yy=M-n
    elif arg0=='z':zz=M-n
#    elif arg0=='w':ww=M-n
    elif arg0=='v':vv=M-n
#    elif arg0=='c':cc=M-n
#    elif arg0=='h':hh=M-n

    if arg1=='x':xx=n
    elif arg1=='y':yy=n
    elif arg1=='z':zz=n
    elif arg1=='v':vv=n
#    elif arg1=='w':ww=n
#    elif arg1=='c':cc=n
#    elif arg1=='h':hh=n

    return vv,xx,yy,zz
```

```python
#################
def rho(*args):
#################
# Calculates the rate of transition between
# homogeneous states based on (14) in the lab script. 
# Pass strategy names as args
  global payoffDict
  assert len(args)==2,'NEED 2 ARGS!!!!'

  strat0=payoffDict[args[0]]
  strat1=payoffDict[args[1]]
  # Look up payoff functions from dictionary i.e. 'x' points to p_x
  returnVal=1.0

  for q in range(1,M): 
    temp=0
    for n in range(1,q+1): 
      vv,xx,yy,zz=getNumbers(args[0],args[1],n)
    # Get number arguments for payoffs
    # Depends on which fixation probability i.e. xy,zx
      temp+=strat0(vv,xx,yy,zz)-strat1(vv,xx,yy,zz)
    returnVal+=np.exp(temp*s)
  return 1.0/returnVal
```

In order to define the full transition matrix, we need to calculate $\rho$ for each pair of strategies

```python
#################
def getTransitionMatrix(strategies): 
#################
  d=len(strategies)
  matrix=np.zeros(shape=(d,d))
  for i in range(d):
    for j in range(d):
      if i==j:
        matrix[i,j]=1-np.sum([(1.0/(d-1))*rho(strategies[i],noti) for noti in strategies if not noti==strategies[i]])
      else: 
        matrix[i,j]=(1.0/(d-1)*rho(strategies[i],strategies[j]))
  return np.array(matrix) 
```

The final function we need will perform the power iteration for us and return the stationary distribution



```python
#################
def getStatDist(mat): 
#################
# Get stationary dist by power method
  tempDist=np.array([0.25]*4)
# Initial distribution is all defectors
  for i in xrange(50000): 
    tempDist=(np.dot(tempDist,mat))
    tempDist/=np.sum(tempDist) 
  return tempDist
```

Then finally we sweep over a range of values of s, the imitation strength in our main function.



```python
#################
def main(): 
#################
  payoffDict['v']=p_v 
  payoffDict['x']=p_x 
  payoffDict['y']=p_y 
  payoffDict['z']=p_z
# Look up payoff functions
  vVals=[]
  xVals=[]
  yVals=[]
  zVals=[]
  sVals=[0.001,0.01,0.1,1,10,100,200,1000,10000]

  global s
# Make s values available to other functions
  for val in sVals: 
    s=val
    mat=getTransitionMatrix(strategies)

    dist=getStatDist(mat)
    print s,dist
    vVals.append(dist[0])
    xVals.append(dist[1])
    yVals.append(dist[2])
    zVals.append(dist[3])
    
  plt.semilogx(sVals,vVals,label=v)
  plt.semilogx(sVals,xVals,label=x)
  plt.semilogx(sVals,yVals,label=y)
  plt.semilogx(sVals,zVals,label=z)
  plt.legend(loc='upper left')
  # Plot proportion of each strategy with a log scaled x axis

  plt.xlabel('Imitation Strength (s)')
  plt.ylabel('Proportion')

  plt.axhline(1.0/len(payoffDict.keys()),linestyle='--',color='k')
  # This line plots equal proportions of each strategy
  plt.show()

if __name__=='__main__': 
  main()
```

Putting everything together yields the following plot. After all that hard work we can see exactly how these strategies perform as we vary the imitation strength. When imitation is very low we see that the agents choose their strategies basically at random, so the proportions are equal.
